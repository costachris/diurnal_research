# Notebook for exploring CMIP6 data provided by Pangeo in the cloud

```python
# necessary url
url = "https://raw.githubusercontent.com/NCAR/intake-esm-datastore/master/catalogs/pangeo-cmip6.json"
# open the catalog
dataframe = intake.open_esm_datastore(url)
```

```python
# models = dataframe.search(experiment_id='historical',
#                           source_id = 'GFDL-CM4')

models_clt = dataframe.search(experiment_id='historical',
                          variable_id = 'lwp')
#                           ember_id = 'r1i1p1f1')
models_clt
```

```python
models_clt.df['source_id'].unique()
```

```python
source_id_dcc_bias = ['FGOALS-g3','GFDL-CM4','GFDL-CM4', 'GFDL-ESM4']
models = models_clt.search(source_id = source_id_dcc_bias[-1])
```
